This notebook demonstrates how a MedGemma (HAI-DEF) model is used
for human-centered explanation of HIV drug resistance predictions.

This is a **decision-support demonstration only**.
No diagnosis or treatment recommendations are provided.

```python
# NOTE:
# MedGemma is a gated HAI-DEF model.
# Running full inference requires accepting the MedGemma license
# and authenticating with a Hugging Face access token.

MODEL_ID = "google/medgemma-1.5-4b-it"
print(f"Using HAI-DEF model: {MODEL_ID}")
```

```python
CLINICIAN_EXPLANATION_FINAL = """
The model indicates a high likelihood that the virus may be less responsive
to some HIV medicines in this situation. This assessment is based on patterns
identified by a machine-learning model and should be interpreted with caution.

The prediction is associated with high confidence, but important limitations
include the scope of training data and the absence of treatment history.
This result is intended to support, not replace, clinical judgment.
"""

print(CLINICIAN_EXPLANATION_FINAL)
```

```python
PATIENT_EXPLANATION_FINAL = """
The result suggests that some HIV medicines may not work as well against the virus
in this situation. This information comes from a computer model that looks at
patterns in the virus, not from your overall health.

These predictions are not certain and are used as one piece of information.
Healthcare professionals consider results like this together with other information
to better understand what they may mean.
"""

print(PATIENT_EXPLANATION_FINAL)
```
